```python
ms_zh_cn = sorted(glob('/home/husein/translation/ms.txt*.splitted.zh-CN.requested'))
len(ms_zh_cn)
```

```python
with open('processed-ms-zh-cn.jsonl', 'w') as fopen_l:
    for f in ms_zh_cn:
        with open(f) as fopen:
            for l in fopen:
                l = json.loads(l)
                try:
                    left = l['src']
                    right = l['r']['result']
                    d = {
                        'left': left,
                        'right': right,
                    }
                    fopen_l.write(f'{json.dumps(d)}\n')
                except:
                    pass
```
