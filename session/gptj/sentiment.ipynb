```python
import json

with open('/home/husein/ssd3/sentiment/explain-sentiment.json') as fopen:
    sentiment = json.load(fopen)
```

```python
json.loads(sentiment[0][1])
```

```python
template = {
    "description": "Template used by Alpaca-LoRA.",
    "prompt_input": "Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.\n\n### Instruction:\n{instruction}\n\n### Input:\n{input}\n\n### Response:\n",
    "prompt_no_input": "Below is an instruction that describes a task. Write a response that appropriately completes the request.\n\n### Instruction:\n{instruction}\n\n### Response:\n",
    "response_split": "### Response:"    
}
```

```python
import random

l = [
    "text `{value}`, categorize the text with label ['negative', 'positive', 'neutral']",
    "teks `{value}`, kategorikan teks dengan label ['negative', 'positive', 'neutral']",
]

r = [
    "text `{value}`, categorize the text with label ['negative', 'positive', 'neutral'] and explain, return as JSON key {{'sentiment', 'explain_en', 'explain_ms'}}",
    "teks `{value}`, kategorikan teks dengan label ['negative', 'positive', 'neutral'] dan terangkan sebab, pulangkan sebagai kunci JSON {{'sentiment', 'explain_en', 'explain_ms'}}",
]
```

```python
data_sentiment = []
for d in sentiment:
    try:
        t = d[0]
        a = d[1]
        j_a = json.loads(d[1])
        if random.random() > 0.7:
            q = random.choice(l)
            a = j_a['sentiment']
        else:
            q = random.choice(r)
            a = a
        
        q = q.format(value=t)
        res = template["prompt_no_input"].format(instruction=q)
        res = f"{res}{a}"
        data_sentiment.append(res)
    except Exception as e:
        print(e)
        pass
```
